####
The depth of closure is the depth beyond which significant sediment transport does not occur. It is a critical concept in coastal engineering for determining the limits of active beach profile changes. This depth is influenced by wave energy and sediment characteristics. 
This code calculates and compares the depth of closure (DoC)‚Äîthe seaward limit of significant sediment movement‚Äîusing several well-known empirical models. The DoC is a key parameter in coastal zone management, beach nourishment design, and shoreline evolution modeling.


‚öôÔ∏è Input Parameters
- H_s: Significant wave height (in meters)
- T_s: Significant wave period (in seconds)
- g: Acceleration due to gravity (9.81 m/s¬≤)

üß™ Methods Implemented
- Hallermeier Inner Limit (1978, 1981)
Estimates DoC based on energetic but non-extreme wave conditions:
d_{inner} = 2.28 H_s - 68.5 \frac{H_s^2}{g T_s^2}- Hallermeier Outer Limit
Represents the maximum depth where sediment transport from storm waves may occur:
d_{outer} = 1.75 H_s- Birkemeier (1985)
Refined for U.S. East Coast field data:
d = 1.57 H_s - 0.005 \frac{H_s^2}{g T_s^2}- Kraus et al. (1998)
Matches the structure of Hallermeier‚Äôs inner formulation, emphasizing long-term changes:
d = 2.28 H_s - 68.5 \frac{H_s^2}{g T_s^2}

üìä Output
The script prints and visualizes the estimated depth of closure from each method using a bar chart. This allows for easy comparison of how different empirical models respond to the same wave inputs.


```python

import numpy as np
import matplotlib.pyplot as plt

# Define the wave parameters
H_s = 2.0# Significant wave height in meters
T_s = 8.0# Significant wave period in seconds
g = 9.81# Acceleration due to gravity in m/s¬≤

# Hallermeier (1978, 1981) Method
def hallermeier_inner(H_s, T_s, g):
    d_inner = 2.28 * H_s - 68.5 * (H_s**2 / (g * T_s**2))
    return d_inner

def hallermeier_outer(H_s):
    d_outer = 1.75 * H_s
    return d_outer

# Birkemeier (1985) Method
def birkemeier(H_s, T_s, g):
    d_birkemeier = 1.57 * H_s - 0.005 * (H_s**2 / (g * T_s**2))
    return d_birkemeier

# Kraus et al. (1998) Method
def kraus(H_s, T_s, g):
    d_kraus = 2.28 * H_s - 68.5 * (H_s**2 / (g * T_s**2))
    return d_kraus

# Calculate depth of closure using different methods
doc_hallermeier_inner = hallermeier_inner(H_s, T_s, g)
doc_hallermeier_outer = hallermeier_outer(H_s)
doc_birkemeier = birkemeier(H_s, T_s, g)
doc_kraus = kraus(H_s, T_s, g)

# Print the results
print(f"Hallermeier Inner Depth of Closure= {doc_hallermeier_inner:.2f} meters")
print(f"Hallermeier Outer Depth of Closure= {doc_hallermeier_outer:.2f} meters")
print(f"Birkemeier Depth of Closure       = {doc_birkemeier:.2f} meters")
print(f"Kraus et al. Depth of Closure     = {doc_kraus:.2f} meters")

# Plot the results
methods = ['Hallermeier Inner', 'Hallermeier Outer', 'Birkemeier', 'Kraus et al.']
doc_values = [doc_hallermeier_inner, doc_hallermeier_outer, doc_birkemeier, doc_kraus]

plt.figure(figsize=(10, 5))
plt.bar(methods, doc_values)
plt.xlabel('Methods')
plt.ylabel('Depth of Closure (meters)')
plt.title('Depth of Closure Calculations Using Different Methods')
plt.xticks(rotation=45)
plt.show()
```
